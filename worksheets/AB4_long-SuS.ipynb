 Die Einträge der User-Matrix sollen nun so gewählt werden, dass die Fehlerfunktion $error_U$ einen minimalen Wert annimmt. Wir lösen  ein  Minimierungsproblem.  Aus der Schule kennst du bereits Minimierungsprobleme dieser Art:

<div class="alert alert-block alert-warning">
<b>Minimierungsproblem mit einer Variablen:</b><p>
Gegeben ist eine Funktion $f$, die von einer Variablen $x$ abhängt. Gesucht ist der Wert für $x$, bei dem die Funktion $f$ mit $f(x)$ einen minimalen Wert annimmt. </div>

In unserem Fall sieht das Problem ähnlich aus. Jedoch lösen wir ein Minimierungsproblem, welches von zwei Variablen abhängt:

<div class="alert alert-block alert-warning">
<b>Minimierungsproblem mit mehreren Variablen:</b><p>
Gegeben ist die Funktion $error_U$, die von den Variablen $u_{11}$ und $u_{12}$ abhängt. Gesucht sind die Werte für $u_{11}$ und $u_{12}$, für die Fehlerfunktion $error_U$ einen minimalen Wert annimmt. </div>




 

Wir übertragen das Vorgehen, welches wir zum Lösen von Minimierungsproblemen mit einer Variablen  kennen, auf das Lösen von Problemen mit mehreren Variablen:

* Bei Funktionen, die von **einer Variablen** abhängen, wird in der Schule meist die Ableitung der Funktion bestimmt und anschließend die Nullstellen der Ableitung berechnet.
* Bei Funktionen, die von **mehreren Variablen** abhängen, kann man  ähnlich vorgehen. Der wesentliche Unterschied ist, dass mehrere **Teilableitungen** berechnet werden müssen. Für jede Variable genau eine.

**Beispiel:**
Die  Ableitung einer Funktion $f$, die von drei Variablen $x_1$, $x_2$ und $x_3$ abhängt, setzt sich aus drei Teilableitungen zusammen. 
$$f'(x_1, x_2, x_3) = \pmatrix{ \text{Ableitung nach} \; x_1 \\ \text{Ableitung nach} \; x_2 \\ \text{Ableitung nach} \; x_3} $$

<div id="information" class="alert alert-success">

        <li>Die Ableitung einer Funktion, die von mehreren Variablen abhängt, wird Gradient genannt.</li>
        <li>Die "Teilableitungen" werden partielle Ableitungen genannt. Sie werden in der Regel als Einträge eines Vektors untereinander notiert.</li>
    </ol>
</div>
