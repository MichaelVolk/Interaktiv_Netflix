Um das Produkt $C = A \cdot B$ zu berechnen gehen wir wie folgt vor:

Der Eintrag $c_{ij}$ wird berechnet, indem man den ersten Eintrag der $i$-ten Zeile von $A$ mit dem ersten Eintrag der $j$-ten Zeile multipliziert. Anschließend werden die zweiten Einträge multipliziert und dazu addiert. Dies wird bis zum Ende der Reihe/Spalte fortgesetzt. (Wir sehen also, warum $A$ so viele Spalten wie $B$ Zeilen braucht: Sonst geht das hier nicht auf).
