### 2) Regularisierung

Wir addieren einen zusätzlichen Term an den Funktionsterm der bisherigen Fehlerfunktion. 
Dieser sogenannte  Regularisierungsterm lautet: 

$$\lambda \cdot \left( u_{11}^2 + u_{21}^2 + u_{31}^2 + u_{21}^2 + u_{22}^2 + u_{23}^2 +  ... +  m_{11}^2 + m_{21}^2 + m_{31}^2 + ...\right)$$


 

Wir lösen damit nun das folgende Minimierungsproblem:
<div class="alert alert-block alert-warning">
<b>Finde die Einträge von U und M, sodass:</b> 
\begin{align}
 \Big(r_{11} - (u_{11} \cdot m_{11} + u_{12} \cdot m_{21})\Big)^2 + \Big(r_{12} - (u_{11} \cdot m_{12} + u_{12} \cdot m_{22})\Big)^2 + ...  {\color{red}{ + \lambda \cdot \left( u_{11}^2 + u_{21}^2 + u_{31}^2 + u_{21}^2 + u_{22}^2 + u_{23}^2 +  ... +  m_{11}^2 + m_{21}^2 + m_{31}^2 + ...\right)}}
\end{align}
<b>minimal wird.</b>
</div>

**Hinweis:** Der Einfluss des addierten Terms kann durch den Parameter $\lambda$ reguliert werden. Für $\lambda$ können wir **beliebige Werte größer gleich 0** festlegen.
