__Wir legen nun noch einen weiteren Eintrag  fest und versuchen die fehlenden Einträge geeignet zu wählen:__ 

 $u_{11} = 2 $, $u_{12} = 0$, $u_{21} = 0$ und  $u_{22} = 2$ 

\begin{align}
&  {\color{red}{5}}  = 2 \cdot m_{11} + 0 \cdot m_{21}\\
& {\color{blue}{3}} = 2 \cdot m_{12} + 0 \cdot m_{22}\\
& {\color{orange}{3}} = 0\cdot m_{11} + 2 \cdot m_{21}\\
& 1 = 0 \cdot m_{12} + 2\cdot m_{22}\\
\end{align}


**Wähle nun die fehlenden vier Einträge von $M$  so, dass die Gleichungen erfüllt sind.**
