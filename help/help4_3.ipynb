_Führe zunächst erneut den Code von Schritt 1 und 2 aus._

#### Schritt 1:

```julia
# Ratingmatrix
R = [3.0 2.0 1.0 4.0;
     1.0 2.0 3.0 4.0]; 

# Zerlegung  
U = [2.0 0.0;
     1.0 1.5];

M = [1.5 1.0 1.0 2.0; 
     0.0 2.0 1.0 2.0];

# Berechnung der Vorhersagematrix:
U * M
```

#### Schritt 2:

```julia
# Berechnung der einzelnen Abweichungen
R - U * M
```

#### Schritt 3: Wie kannst du die Einträge nun zu einem einzigen Fehlerwert zusammenrechnen?

Du kannst die Abweichungen zwischen der Ratingmatrix `R` und der Vorhersage `U*M` aufsummieren. Führe den folgenden Code aus.

```julia
sum(R-U*M)
```

Aber... wenn wir einfach alle Abweichungen ausummieren, heben sich positive und negative Abweichungen auf. 
Um das zu verhindern kann man veschiedene Ideen verfolgen.

#### Idee 1:  Betrag der einzelnen Abweichungen aufsummieren:

```julia
# Betrag der einzelnen Einträge
abs.(R-U*M)
```

```julia
# Summe der Beträge
sum(abs.(R-U*M))
```


#### Idee 2:  Das Quadrat der einzelnen Abweichungen aufsummieren:


```julia
# Quadrieren der einzelnen Einträge
(R-U*M).^2
```

```julia
# Summe der quadratischen Abweichungen
sum((R-U*M).^2)
```
