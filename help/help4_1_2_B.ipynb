Du kannst  die Differenz zwischen jedem einzelnen Eintrag  der Rating-Matrix `R` und der Vorhersage-Matrix `P` berechnen und aufsummieren. Dazu kannst du im Code schreiben:

`(r11-(u11*m11+u12*m21)) + (r12-(u11*m12+u12*m22)) + (r21-(u21*m11+u22*m21)) + (r22-(u21*m12+u22*m22))`

Überlege dir nun noch, wie du diese Formel abändern musst, damit in der Summe keine negativen Abweichungen vorkommen können. Nutze bei der Eingabe folgende Hinweise: 


**<i class="fa fa-exclamation-triangle" style="font-size:22px"></i>  Hinweise: Julia als Taschenrechner nutzen**

 
* Quadrieren: `4^2`
* Die Wurzel ziehen: `sqrt(16)`
* Den Betrag einer Zahl berechnen: `abs(-4)`
