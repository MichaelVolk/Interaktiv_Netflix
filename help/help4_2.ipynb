  2 & 0   \\

 M = \begin{pmatrix} 
 1.5 & 1 & 1 & 2 \\

#### Schritt 1: Berechnung der "Vorhersagematrix"  
____
<i class="fa fa-laptop" style="font-size:38px"></i>

_Berechne das Produkt `U*M`. Dies liefert dir die Vorhersagematrix `P`. Führe den folgenden Code aus um  das Produkt zu berechnen._

```julia
# Ratingmatrix
R = [3.0 2.0 1.0 4.0;
     1.0 2.0 3.0 4.0]; 

# Zerlegung  
U = [2.0 0.0;
     1.0 1.5];

M = [1.5 1.0 1.0 2.0; 
     0.0 2.0 1.0 2.0];

# Berechnung der Vorhersagematrix:
U * M
```

#### Schritt 2: Berechnung der Abweichung zwischen Vorhersagematrix und Ratingmatrix

____
<i class="fa fa-laptop" style="font-size:38px"></i>

Du kannst zudem berechnen, wie stark die einzelnen Einträge der Vorhersagematrix von der tatsächlichen Ratingmatrix abweichen. Führe den folgenden Code aus. 

```julia
# Berechnung der einzelnen Abweichungen
R - U * M
```
