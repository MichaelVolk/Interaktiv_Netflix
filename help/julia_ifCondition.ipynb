# if-Bedingung

### Motivation
Angenommen wir wollen eine Berechnung oder Anweisung nur dann duchführen, wenn eine bestimmte Bedingung erfüllt ist. Dann könnten wir das mit einer sogenannten if-Bedingung tun.

### Formal:
Werfen wir zuerst einen Blick auf die formale Syntax:

```julia
if Bedingung I 
    Anweisung(en) I
    
elseif Bedingung II
    Anweisung(en) II

elseif Bedingung III
    Anweisung(en) III
    
 ... usw...
    
elseif Bedingung X
    Anweisung(en) X
end
```
In Worten kann man die if-Bedingung so beschreiben:
* In der ersten Zeile wird die 1. Bedingung eingegeben.
* Ist Bedingung I  erfüllt? 
    * JA:   Anweisung I wird ausgeführt. STOP
    * NEIN: Anweisung I wird nicht ausgeführt. Weiter gehts zu Bedingung II.
* Ist Bedingung II  erfüllt?  
    * JA:   Anweisung II wird ausgeführt. STOP
    * NEIN: Anweisung II wird nicht ausgeführt. Weiter gehts zu Bedingung III.
* wenn bisher keine Bedingung erfüllt wurde, werden nach und nach alle weiteren Bedingungen durchgegangen. 

**Hinweis:** Schreibt man `else` anstelle des letzten `elseif`, so führt Julia die zugehörige Anweisung aus, wenn alle bisherigen Bedingungen nicht erfüllt waren.

### Beispiel 1:
* Der Wert der Variable `a` soll quadriert werden,  **wenn** `a` negativ ist. 
* Der Wert der Variable `a` soll halbiert werden,  **wenn** `a` positiv ist. 
* Der Wert der Variable `a` soll um 12 erhöht werden,  **wenn** `a` positiv ist. 

(*Klicke in das folgende Code-Feld und drücke auf "Run".*)

````julia
# Speichern eines Wertes für die Variable a
a = -2; # Ändere den Wert von a und beobachte das Ergebnis.

if a < 0 
       a = a^2;
    
    elseif a == 0
       a = a + 12;
    
    else # Hier hätten wir auch elseif a > 0 schreiben können
       a = a / 2;
end
    
println("Der Wert von a ist $a")
````

### Beispiel 2: Kombination von for - Schleife und if - Bedingung
In dem Vektor `positions` sind viele mögliche Positionen gespeichert.
* Ist ein Eintrag in dem Vektor `positions` negativ, so soll dieser Eintrag zu dem Vektor `vNeg` hinzugefügt werden.
* Andernfalls soll der Eintrag dem Vektor `vPos` hinzugefügt werden.

(*Klicke in das folgende Code-Feld und drücke auf "Run".*)

````julia
# Definieren der Vektoren
vNeg  = [] # Wir starten mit einem leeren Vektor
vPos  = [] # Wir starten mit einem leeren Vektor

positions = [3,-4,0,1,-1,-3,2,-2,4] # Hier speichern wir alle Positionen
nPositions = size(positions,1) # Hier bestimmen wir die Länge des Vektors `positions`. Dies ist gerade die Anzahl der Einträge des Vektors.

for i = 1 : nPositions # Für jeden Eintrag in dem Vektor `positions` ...
    
    if positions[i] < 0 # prüfen wir nach und nach, ob der Wert an der Stelle i negativ ist.
        vNeg = push!(vNeg,positions[i]) # Ist der Wert negativ, so wird der Eintrag zu dem Vektor vNeg hinzugefügt.
    else
        vPos = push!(vPos,positions[i]) # Ist der Wert null oder positiv, so wird der Eintrag zu dem Vektor vPos hinzugefügt.
    end

end

println("Das Ergebnis sind die beiden Vektoren vNeg =  $vNeg und vPos = $vPos");
````
