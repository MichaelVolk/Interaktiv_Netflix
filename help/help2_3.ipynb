Hier findest du eine Beispielrechnungen, für die orange markierte Bewertung in der Ratingmatrix:

${\color{orange}{r_{45}}} = u_{4A} \cdot m_{A5} + u_{4C} \cdot m_{C5}$

Die Eingabe im Code würde lauten: `u4A * mA5 + u4C * mC5`
___
Übertrage diese Rechnung auf die Aufgabe auf dem Arbeitsblatt.
