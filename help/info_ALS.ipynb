# Das Optimierungsverfahren

Auf diesem Arbeitsblatt entwickeln wir das Optimierungsverfahren **Alternierende Methode kleinster Fehlerquadrate**.   
Bei diesem Verfahren werden **abwechselnd** alle Einträge  der User-Matrix und dann alle Einträge der Movie-Matrix optimiert:

**Optimierung der User-Matrix:**

1. **Schritt:** Die Einträge der Movie-Matrix werden auf beliebige Werte festgelegt.  $M$  <i class="fas fa-thumbtack"></i>
2. **Schritt:** Die Einträge der User-Matrix werden variiert. Wir  wählen die Einräge von $U$ so, dass der Fehler minimal wird.  

**Optimierung der Movie-Matrix:**  

3. **Schritt:** Die gefundenen Einträge der User-Matrix werden festgehalten. $U$ <i class="fas fa-thumbtack"></i>
4. **Schritt** Jetzt werden die Einträge der Movie-Matrix variiert.  Wir  wählen die Einräge von $M$ so, dass der Fehler minimal wird.

**Optimierung der User-Matrix:**

5. **Schritt:** Die gefundenen Einträge der Movie-Matrix werden   festgehalten. $M$  <i class="fas fa-thumbtack"></i>
6. **Schritt**  Die Einträge der User-Matrix werden optimiert. Wir  wählen die Einräge von $U$ so, dass der Fehler minimal wird.
