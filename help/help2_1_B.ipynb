Hier findest du **Beispielrechnungen** für die Bewertungen, die schon eingetragen sind:

* Die Bewertung von **User 2 für Film 1** ergibt sich aus **Zeile 2** der User-Tabelle und  **Spalte 1** der Movie-Tabelle:  
$5 = 1 \cdot 3 + 1 \cdot 2$

* Die Bewertung von **User 1 für Film 3** ergibt sich aus   **Zeile 1** der User-Tabelle und  **Spalte 3** der Movie-Tabelle:  
$4= 0 \cdot 1 + 1 \cdot 4$

* Die Bewertung von **User 4 für Film 4** ergibt sich aus   **Zeile 4** der User-Tabelle und   **Spalte 4** der Movie-Tabelle:  
$3 = 1 \cdot 1 + 0.5 \cdot 4$



**Übertrage diese Rechnung auf die fehlenden Bewertungen!**
