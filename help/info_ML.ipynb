## Infoblatt | Unterteilung der Daten in Trainings- und Testdaten

Um zu bewerten, wie gut sich unser Vorgehen für unbekannte Daten eignet, nutzen wir eine Strategie aus dem Bereich des **Maschinellen Lernens**. Dabei werden die bekannten Bewertungsdaten unterteilt. Der eine Teil der Daten wird dann zum "Lernen" des Modells genutzt, der andere Teil der Daten wird genutzt, um   anschließenden  zu überprüfen wie gut das Modell für Daten geeignet ist, die nicht zum Lernen des Modells genutzt wurden. 

Die wesentlichen Schritte, die auch in der folgenden Abbildung dargstellt sind, lauten:

1. **Schritt: Unterteilung der Daten in Trainings- und Testdaten**

2. **Schritt: Trainingsphase:**
    * Wir bestimmen für die Trainingsmatrix ein Zerlegung in $U$ und $M$. 
    * Die gefundene Zerlegung nutzen wir, um die  Vorhersage-Matrix zu berechnen.
  
    
3. **Schritt: Testphase:**
    * Wir vergleichen die **vorhergesagten  Ratings**  mit den **tatsächlichen Ratings** aus der Testmatrix und
    * bewerten, wie gut unsere Zerlegung für die Vorhersage von **unbekannten Ratings** geeignet ist. Dazu  berechnen wir der Fehler für die Testdaten (Testfehler) mit unserem bereits definierten Fehlermaß.
