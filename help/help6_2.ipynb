# Tipp 2 | Partielle Ableitungen berechnen


#### Berechnung der Teilableitung nach $u_{11}$:
Wir berechnen die Ableitung der Funktion 
$$error(u_{11}, u_{12}) = (2 {\color{blue}{-}} (u_{11} + u_{12}))^{\color{red}{2}} + (3 {\color{blue}{-}}  (u_{11} + u_{12}))^{\color{red}{2}} $$
nach $u_{11}$. Dazu behandeln wir die Variable  $u_{12}$  wie eine Konstante. 


#### Lösung:

Wir wenden die Kettenregel an. Das Ergebnis für die **Teilableitung  nach $\;u_{11}$** lautet dann:


$${\color{red}{2}} \cdot(2-u_{11}-u_{12}) \cdot {\color{blue}{(-1)}} + {\color{red}{2}}  \cdot (3-u_{11}-u_{12}) \cdot {\color{blue}{(-1)}}$$

Versuche das Ergebnis nachzuvollziehen und bestimme selbst die **Teilableitung  nach $\;u_{12}$**!
