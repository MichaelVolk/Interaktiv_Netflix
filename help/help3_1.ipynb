Damit die User-Matrix `U1` und die Movie-Matrix `M1` die Rating-Matrix `R1` korrekt widerspiegeln, müssen deine Formeln von Arbeitsblatt 2 stets die richtige Bewertung liefern.

 


 Wir erhalten ein Gleichungssystem mit **4 Gleichungen mit 8 Unbekannten**:


\begin{align}
&   {\color{red}{5}}   = u_{11} \cdot m_{11} + u_{12} \cdot m_{21} & &\text{(1. Zeile von $U1$ und 1. Spalte von $M1$)}\\
& {\color{blue}{3}} = u_{11} \cdot m_{12} + u_{12} \cdot m_{22} & &\text{(1. Zeile von $U1$ und 2. Spalte von $M1$)}\\
& {\color{orange}{3}} = u_{21} \cdot m_{11} + u_{22} \cdot m_{21} & &\text{(2. Zeile von $U1$ und 1. Spalte von $M1$)}\\
& {\color{black}{1}} = u_{21} \cdot m_{12} + u_{22} \cdot m_{22}& & \text{(2. Zeile von $U1$ und 2. Spalte von $M1$)}\\
\end{align}


 


Lege einzelne Einträge von $U$ oder $M$ auf beliebige Werte fest. Versuche   die übrigen Einträge so zu wählen, dass alle Gleichungen erfüllt sind.

#### Beispiel:
Starte mit $u_{11} = 2 $, $u_{12} = 0$ und $u_{21} = 0$: 

\begin{align}
&  {\color{red}{5}}  = 2 \cdot m_{11} + 0 \cdot m_{21}\\
& {\color{blue}{3}} = 2 \cdot m_{12} + 0 \cdot m_{22}\\
& {\color{orange}{3}} = 0\cdot m_{11} + u_{22} \cdot m_{21}\\
& 1 = 0 \cdot m_{12} + u_{22} \cdot m_{22}\\
\end{align}


**Wähle die fehlenden Einträge so, dass die Gleichungen erfüllt sind!**
