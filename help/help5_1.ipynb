# Tipp 1 | Bestimmung des Minimums

Wir müssen das folgende lineare Gleichungsystem lösen:

$\text{I.  Ableitung nach } u_{11} = 0 \\
 \text{II.  Ableitung nach } u_{12} = 0  $

Setze die bereits berechneten Teilableitungen ein und löse das Gleichungssystem nach $u_{11}$ und $u_{12}$ auf.

**Hinweis:** das Gleichungssystem ist in diesem Fall **unterbestimmt**. Das heißt, es gibt unendlich viele Lösungen. Du kannst eine beliebige Lösung wählen.
