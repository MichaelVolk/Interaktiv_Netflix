Um nun auch geometrisch begründen zu können, wann das Ergebnis eine größere oder kleinere Bewertung ist, brauchst du noch die Defintion des Skalarprodukts zweier Vektoren.

#### Definition:

Das Skalarprodukt von zwei Vektoren $\vec{a}$ und $\vec{b}$ ist im zweidimensionalen Fall definiert als:

$\vec{a} \cdot \vec{b} = \cos(\alpha) \cdot (|\vec{a}| \cdot |\vec{b}|)$

Dabei ist $\alpha$ der Winkel den die beiden Vektoren $\vec{a}$ und $\vec{b}$ einschließen.
