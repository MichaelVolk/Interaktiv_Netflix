Um z. B. die Differenz zwischen vorhergesagter und richtiger Bewertung von User 1 für Film 1 zu berechnen, kannst du im Code schreiben:

`(r11-(u11*m11+u12*m21))`

**<i class="fa fa-exclamation-triangle" style="font-size:22px"></i>  Achtung:** Wenn wir nur die Differenz bilden und diese aufsummieren, kann das Ergebnis auch negativ sein. Überlege dir, wie du dies verhindern kannst. Nutze bei der Eingabe deiner Formel die  **Hinweise zu Julia** vom Arbeitsblatt.
