* E1 = Eigenschaft 1 
* E2 = Eigenschaft 2 

Hier findest du eine Beispielrechnungen, für die orange markierten Bewertung in der Ratingmatrix:

${\color{orange}{r_{44}} = u_{41} \cdot m_{14} + u_{42} \cdot m_{24} }$  

Diese Formel kann man so lesen:

$\textbf{(Wie sehr mag User 4 Eigenschaft 1)} \cdot \textbf{(Wie viel von Eigenschaft 1 hat Film 4)} + \textbf{(Wie sehr mag User 4 Eigenschaft 2)} \cdot \textbf{(Wie viel von Eigenschaft 2 hat Film 4)}$

Die Eingabe im Code würde lauten: `u41 * m14 + u42 * m24`
___
Übertrage diese Rechnung auf die Aufgabe auf dem Arbeitsblatt.
