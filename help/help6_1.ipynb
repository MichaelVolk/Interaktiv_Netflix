# Tipp 1 | Partielle Ableitungen berechnen


**Beispiel:**
Unsere  Funktion $error$ hängt von den beiden Variablen $u_{11}$, $u_{12}$ ab:

$$error(u_{11}, u_{12}) = (2 - (u_{11} + u_{12}))^2 + (3 - (u_{11} + u_{12}))^2$$
Wir müssen also zwei Teilableitungen berechnen: 

$$error'(u_{11}, u_{12}) = \pmatrix{ \text{Ableitung nach} \;u_{11} \\ \text{Ableitung nach} \; u_{12}} $$


* Wenn wir die Ableitung nach $u_{11}$  berechnen, behandeln wir die andere Variable (also $u_{12}$) wie eine Konstante (= feste Zahl).
* Wenn wir die Ableitung nach $u_{12}$  berechnen, behandeln wir die andere Variable (also $u_{11}$) wie eine Konstante (= feste Zahl).

**Hinweis:** Zur Berechnung der Teilableitungen benötigst du die Kettenregel!
